```python
from astropy.cosmology import WMAP9 as cosmow
from astropy.cosmology import Planck18 as cosmop
from astropy.constants import L_sun
import numpy as np
import matplotlib.pyplot as plt
```

```python
z=np.linspace(0.1, 500, 100)
def flux(lumdist):
    lsun = L_sun.cgs.value
    return lsun/(4.0*np.pi*(lumdist**2))


plt.plot(z, ((flux(cosmow.luminosity_distance(z).value) - flux(cosmop.luminosity_distance(z).value))/(flux(cosmop.luminosity_distance(z).value)))*100)
plt.xscale('log')
```

```python
z=np.linspace(0, 500, 1000)
plt.plot(z, ((cosmow.age(z).value - cosmop.age(z).value)/(cosmop.age(z).value))*100)
plt.xscale('log')
```
