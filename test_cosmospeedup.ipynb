```python
import numpy as np
import matplotlib.pyplot as plt
import lbg_forecast.popmodel as pop
from astropy.cosmology import WMAP9 as cosmo
```

```python
def tuniv_grid(zgrid):
    tuniv = []
    for z in zgrid:
        tuniv.append(cosmo.age(z).value)

    return np.asarray(tuniv)
```

```python
def get_tuniv(z, zgrid, tuniv):
    return tuniv[np.where(zgrid == z)]
```

```python
z_grid = np.linspace(0, 15, 1000)
t_univ = tuniv_grid(z_grid)
```

```python
get_tuniv(z_grid[-5], z_grid, t_univ)
```

```python
plt.plot(z_grid, t_univ)
plt.plot(np.linspace(0, 15, 10), np.interp(np.linspace(0, 15, 10), z_grid, t_univ))
```
