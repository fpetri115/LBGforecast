```python
import numpy as np
import matplotlib.pyplot as plt
import lbg_forecast.nz_sim as nz
```

```python
def gather_redshifts(nfiles, path):
    i = 0
    nz_data = []
    while(i < nfiles):
        nzs = np.load(path+"/emulated_redshifts_"+str(i)+".npy", allow_pickle=True)
        nsamples = nzs.shape[0]
        j = 0
        while(j < nsamples):
            nz_data.append(nzs[j, :])
            j+=1
        i+=1

    return np.asarray(nz_data)
```

```python
def plot_multiple_realisations(redshifts, min, max, alpha, density):

    dz = 0.1
    minz = 0.0
    maxz = 7.0
    bins = np.arange(minz, maxz, dz)

    i = min
    while(i <= max):
        nz.visualise_redshifts(redshifts[i], bins=bins, alpha=alpha, density=density)
        i+=1
```

```python
nzs = gather_redshifts(120, "./redshifts")
total_nzs = len(nzs)
print("total nzs:", total_nzs)
```

```python
plot_multiple_realisations(nzs, 0, 1000, alpha=0.01, density=True)
```
